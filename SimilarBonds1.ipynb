```python
"""Normalise - however without normalisation it still gives the same answares
"""
A = preprocessing.StandardScaler().fit(X).transform(X)  #
```

```python
B=A
```

```python
d=cdist(A,B,metric='euclidean') 
```

```python

"""Loop through the diagonals from 0 up. 
   Please note I have not tested too much of below resoluts, this is just an implementation of an idea
"""

min1=10000
for c in range(len(d)):
    if c>0:
        d1=d.diagonal(c)
        if (min1>d1.min() and d1.min()>0):
            min1=d1.min()
            coord=np.argmin(d1)
            d1min=d1
            cmin=c

print(min1)
#print(d1min)
coordA=coord
coordB=coord+cmin
print(coordA)
print(coordB)            
```
